Handling the missing values

 Convert all -200 to NaN

 Replace all NAN values with mean of that specific column

```python
air_quality_data = air_quality_data.replace(to_replace = -200, value = np.NaN)
```

```python
air_quality_data.isnull().sum()
```

```python
air_quality_data.tail()
```

```python
air_quality_data.select_dtypes(include=['number']).mean()
```

```python
#Replacing the missing values with the mean value of each column
air_quality_data = air_quality_data.fillna(air_quality_data.select_dtypes(include=['number']).mean())
```

```python
#Converting the date from dd-mm-yyy to yyy-mm-dd as per algorithm
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info = air_quality_data['Time']
time_info
```

```python
#Converting date from h.m.s to h:m:s as per documentation
time_info = time_info.apply(lambda x:x.replace('.', ':'))
print(time_info)
```

```python
print(type(date_info))
```

```python
print(type(time_info))
```

```python
#Combining 2 series to a pandas df
date_time = pd.concat([date_info, time_info], axis = 1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
#combining date and time to one column
date_time['ds'] = date_time['Date'].astype(str)+''+date_time['Time'].astype(str)
```

```python
date_time.info()
```

converting ds dtype from object to date time format

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'], format='%Y-%m-%d%H:%M:%S')
print(data['ds'])
```

```python
from prophet import Prophet
```

```python
#training the prophet model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq ='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail() #yhat is predicted value

```

```python
fig1 = model.plot(forecast)

```

```python
fig2 = model.plot_components(forecast)

```
