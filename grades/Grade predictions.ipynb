```python
import pandas as pd
```

```python
grade_path = r'C:\Users\HP\ML PARCTICE\student-por.csv'
```

```python
grade_data = pd.read_csv(grade_path, sep=";")
grade_data
```

```python
grade_data.describe()
```

```python
grade_data.head()
```

```python
grade_data.columns
```

```python
features= ['failures', 'studytime', 'absences', 'G1', 'G2', 'G3']
```

```python
X = grade_data[features]
```

```python
X.head()
```

```python
y = grade_data.G3
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size = 0.1, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
grade_model = DecisionTreeRegressor(random_state = 1)
```

```python
grade_model.fit(train_X,train_y)
```

```python
predictions = grade_model.predict(test_X)
```

```python
print (predictions)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(test_y, predictions)
```

```python
print(mae)
```
