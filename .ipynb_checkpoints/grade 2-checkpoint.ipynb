```python
import pandas as pd
```

```python
grade_path = r'C:\Users\HP\ML PARCTICE\student-mat.csv'
```

```python
grade_data = pd.read_csv(grade_path, sep=";")
grade_data
```

```python
grade_data.columns
```

```python
features =['absences', 'studytime', 'failures', 'G1',  'G2',  'G3']
```

```python
features
```

```python
X= grade_data[features]
```

```python
X
```

```python
y = grade_data. G2
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split (X, y, random_state = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
grade_model =RandomForestRegressor (random_state = 1)
```

```python
grade_model.fit(train_X, train_y)
```

```python
predictions = grade_model.predict(val_X)
predictions
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(predictions, val_y)
```

```python
print(mae)
```
